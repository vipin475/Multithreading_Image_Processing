```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = open('Report/report','r',encoding='UTF-8')
```

```python
threads = []
wall_time = []
```

```python
for line in data:
    inst = [i for i in line.split()]
    threads.append(int(inst[0]))
    wall_time.append(float(inst[1]))
```

```python
data.close()
```

```python
plt.xticks(np.arange(0, 11, 1)) 
plt.yticks(np.arange(min(wall_time), max(wall_time), 10)) 


plt.xlabel('Number of threads')
plt.ylabel('Wall clock Time (sec)')

plt.xticks(np.arange(0, 11, 1))
plt.yticks(np.arange(int(min(wall_time)), int(max(wall_time))+101, 100))

plt.plot(threads,wall_time,marker='o',color='g',label='wall_time')

for x,y in zip(threads,wall_time):
    label = "{:.2f}".format(y)
    plt.annotate(label, (x,y), textcoords="offset points",xytext=(0,10), ha='center',fontsize=8)

plt.legend()
plt.plot()
```

```python
multithreading = open('Report/Using_10_thread','r',encoding='UTF-8')
no_multithreading = open('Compare/report_no_multithreading','r',encoding='UTF-8')
```

```python
files = []
wall_time1 = []
wall_time2 = []
```

```python
for line in no_multithreading:
    inst = [i for i in line.split()]
    files.append(int(inst[0]))
    wall_time1.append(float(inst[1]))

for line in multithreading:
    inst = [i for i in line.split()]
    wall_time2.append(float(inst[1]))
```

```python
no_multithreading.close()
multithreading.close()
```

```python
plt.figure(figsize=(15, 8))
plt.title("Wall time")
plt.xlabel('Number of Files')
plt.ylabel('Wall clock Time (sec)')

plt.plot(files,wall_time1,marker='o',color='g',label='no multithreading')
plt.plot(files,wall_time2,marker='o',color='b',label='multithreading')

for x,y in zip(files,wall_time1):
    label = "{:.2f}".format(y)
    plt.annotate(label, (x,y), textcoords="offset points",xytext=(0,10), ha='center')

for x,y in zip(files,wall_time2):
    label = "{:.2f}".format(y)
    plt.annotate(label, (x,y), textcoords="offset points",xytext=(0,10), ha='center')

plt.legend()
plt.show()
```
